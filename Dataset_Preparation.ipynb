```python
import os
import numpy
directory = os.getcwd() + '/Dataset/labels'
```

```python
songs_list = []
for file in os.listdir(directory):
    if file != '.DS_Store':
        for category in os.listdir(directory+'/'+file):
            lst = [category]
            for line in open(directory+'/'+file+'/'+category):
                lst.append(line[:-1])
            songs_list.append(lst)
print(songs_list)
```

```python
for song_list in songs_list:
    save_directory = os.getcwd() + '/Dataset/' + song_list[0][:-4]
    find_directory = os.getcwd() + '/Dataset/lpd_5/lpd_5_cleansed'
    try:
        os.mkdir(save_directory)
    except:
        continue
    print(os.listdir(find_directory+'/B/S/D/TRBSDJW128F92D6E19'))
    print('moving files from', song_list[0])
    for song in song_list[1:]:
        for name in os.listdir(find_directory+'/'+song[2]+'/'+song[3]+'/'+song[4]+'/'+song):
            try:
                os.rename(find_directory+'/'+song[2]+'/'+song[3]+'/'+song[4]+'/'+song+'/'+name, save_directory+'/'+song+name)
            except:
                continue
```
